### Importing Pymongo connector to connect to MongoDb and store tweets in it

```python
from pymongo import MongoClient
import json
```

Below loop uses twitter search api to search based on the ticker symbols or stock symbols. Then stores it in the mondo db.

Additionally, I also added wait times to ensure the search api does not reach twitter limits. (Twitter allows 900 requests per 15 minute time inteval)

```python
i=0
MONGO_HOST = 'mongodb://localhost/stock_tweets'
client = MongoClient(MONGO_HOST)
db = client.stock_tweets
import time

for stock in ['LMND','ZM','OKTA','FVRR','FLGT','TSLA','AAPL','BAC']: #
    
    delay= 60*5  #5 minutes
    t_end = time.time() + delay
    i=1
    while ((time.time() < t_end)):
        q = "("+stock+") since:2021-05-05 until:2021-05-06 lang:en"
        for tweet in tweepy.Cursor(api.search, q=q,result_type='mixed',count=100,lang="en",tweet_mode='extended').items():
            status = api.get_status(tweet.id, tweet_mode="extended")
            print(i)
            i+=1

            try:

                data = {"created_at":str(tweet.created_at),"stock": stock, "tweet":status.retweeted_status.full_text }
                datajson = json.loads(json.dumps(data))
              

                print('Tweet collected at ' + str(data["created_at"]))
                db.twitter_search.insert_one(datajson)
                print(datajson)

            except AttributeError: # Not a Retweet
                data = {"created_at":str(tweet.created_at),"stock": stock, "tweet":status.full_text }
                datajson = json.loads(json.dumps(data))
               

                print('Tweet collected at ' + str(data["created_at"]))
                db.twitter_search.insert_one(datajson)
                print(datajson)

            print("===============================================================")
            if(i%150 == 0):
                
                time.sleep(60*0.5)
                
            if(i>890):
                break
    time.sleep(60*16)

```
